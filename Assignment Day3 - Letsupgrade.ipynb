# Assignment 1
Use IF ELSE and ELIF to write a program in python for your Report Cards.

```python
grade=input("Enter your marks")
if grade >=95:
    print("Your grade is A+")
elif grade>=85 and grade<=94:
    print("Your grade is A")
elif grade>=75 and grade<=84:
    print("Your grade is B+")
elif grade>=65 and grade<=74:
    print("Your grade is B")
elif grade>=55 and grade<=64:
    print("Your grade is C+")
elif grade>=45 and grade<=54:
    print("Your grade is C")
else:
    print("You have failed!")
```

# Assignment 2
Use For Loop to Print Prime Numbers in between 1 to 1000

```python
for i in range(1,1001):
    count=0
    for j in range(1,i/2+1):
        if(i%j==0):
            count=count+1
    if count<2 and i!=1:
        print(str(i)+" is a prime number")
```

# Assignment 3
Write a program for printing the tables from 1,10 using Nested For Loop

```python
for i in range(1,11):
    print("Table of "+str(i))
    for j in range(1,11):
        print(str(i)+"*"+str(j)+"="+str(i*j))
```
